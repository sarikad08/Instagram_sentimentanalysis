```python
import re
import pandas as pd
from InstagramAPI import InstagramAPI
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# List of target usernames to extract comments from
target_usernames = ['ramnavami_pune', 'shree_ram_navmi_bsn']

username = 'dsarika6901'
password = 'project_T'

comments_list = []
sentiment_list = []

# Initialize the Vader sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

for target_username in target_usernames:
    api = InstagramAPI(username, password)
    api.login()

    api.searchUsername(target_username)
    user_id = api.LastJson['user']['pk']

    api.getUserFeed(user_id)
    media = api.LastJson['items'][:400] # get the latest 80 posts

    for post in media:
        media_id = post['id']
        api.getMediaComments(media_id)
        comments = api.LastJson['comments']
        for comment in comments:
            text = comment['text']
            # Use regex to remove any mentions of usernames
            cleaned_text = re.sub(r'@[a-zA-Z0-9._]+', '', text)
            print(cleaned_text)
            # Perform sentiment analysis using Vader
            sentiment_scores = analyzer.polarity_scores(cleaned_text)
            # Assign a sentiment label based on the compound score
            if sentiment_scores['compound'] >= 0.05:
                sentiment = 'positive'
            elif sentiment_scores['compound'] <= -0.05:
                sentiment = 'negative'
            else:
                sentiment = 'neutral'
            comments_list.append(cleaned_text)
            sentiment_list.append(sentiment)
            print(f"Comment: {cleaned_text}\nSentiment: {sentiment}\n")

# Create a pandas DataFrame from the list of comments and sentiment labels
df = pd.DataFrame({'comments': comments_list, 'sentiment': sentiment_list})

# Save the comments and sentiment labels to an Excel file
df.to_excel('comments.xlsx', index=False)
```
